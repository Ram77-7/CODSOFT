Applying Elbow technique

```python
k_rng=range(1,10)
sse=[]
for k in k_rng:
  km=KMeans(n_clusters=k)
  km.fit(df[['petal_length','petal_width']])
  sse.append(km.inertia_)
```

```python
sse
```

sse -define's that sum of square errors.

used to measure discrepancy between observed data and the values predicted by  a model.

```python
plt.xlabel('k_rng')
plt.ylabel('sum of squared error')
plt.plot(k_rng,sse)
```

Optimized cluster at 3 point.

Apply KMean Algorithm

it is used to cluster the partions of data set based on feature similarity.

```python
km=KMeans(n_clusters=3,random_state=0)
y_predicted=km.fit_predict(df[['petal_length','petal_width']])
y_predicted
```

here assigned random_state value as 0 because it takes no new values  or new changes in dat set in every iterations.

```python
df['cluster']=y_predicted
df.head(150)
```

Accuracy Measure

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(df.species,df.cluster)
cm
```

by using species and cluster coloumn's i used to construct confustion matrix .

```python
true_labels=df.species
predicted_labels=df.cluster

cm=confusion_matrix(true_labels,predicted_labels)
class_labels=['Setosa','versicolor','virginica']


#plot Confustion matrix
plt.imshow(cm,interpolation='nearest',cmap=plt.cm.Greens)
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks=np.arange(len(class_labels))
plt.xticks(tick_marks,class_labels)
plt.yticks(tick_marks,class_labels)


#Fill matrix with values
for i in range(len(class_labels)):
  for j in range(len(class_labels)):
    plt.text(j,i,cm[i][j],ha='center',va='center',color='white')

plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()
```

here we get the clear visualiztion for better understanding confustion matrix classification.
